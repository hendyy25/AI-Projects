### Price Prediction Exercise

**Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020**

```python
from sklearn import linear_model
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
year = df.drop('per capita income (US$)', axis = 'columns')
year # dataframe 
```

```python
capita_income = df['per capita income (US$)']
capita_income # pandas series
```

```python
# plot to have an intitial sense of the correlation 
plt.plot(df.year, capita_income, marker = '*', linestyle='')
plt.xlabel('year')
plt.ylabel('capita income')
```

```python
reg = linear_model.LinearRegression()
reg.fit(year, capita_income) # fit(dataframe, pandas series)
```

```python
reg.predict([[2020]])
```
