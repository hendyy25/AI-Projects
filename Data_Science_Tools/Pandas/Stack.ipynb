### Dataframe Reshaping Using stack/unstack

```python
import pandas as pd 
df = pd.read_excel('stocks.xlsx',header = [0,1])
df
```

```python
new_df = df 
new_df = new_df.stack()

new_df.to_csv('Company.csv')
```

```python
df.stack(level=0)
```

```python
stacked_df = df.stack()
print(stacked_df)
```

```python
stacked_df.unstack()
```
