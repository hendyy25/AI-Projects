```python
# Blur
blur = cv.GaussianBlur(img, (7,7), cv.BORDER_DEFAULT) # the tuple (3,3) take only odd numbers, and both of them must be equall, their value reflects the intensity of the blur 
cv.imshow('Blur', blur)
cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
```

```python
# Edge Cascade 
canny = cv.Canny(blur,125, 175)
cv.imshow('Canny Edges', canny)
cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
# uself for medical detections 
```

```python
# Dilating the image 
dilated = cv.dilate(canny, (7,7), iterations = 3)
cv.imshow('Dilated',dilated)
cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
# continuously increases the boundaries of regions of foreground pixels 
```

```python
eroded = cv.erode(dilated, (7,7),iterations =3)
cv.imshow('Eroded', eroded)
cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
# it erodes away the boundaries of foreground object (Always try to keep foreground in white)
```

### Reading Videos 

```python
capture = cv.VideoCapture('../Resources/Videos/dog.mp4')
counter = 0 
frames = []

while True:
    is_true, frame = capture.read() # if there is a frame, return True and that frame
    if is_true:
        frames.append(frame)
        counter+=1
        cv.imshow('my video', frame)
        if cv.waitKey(10) and 0xFF == ord('d'):
            break
    else:
        break

capture.release() # 1) release software resource, 2) release hardware resource
cv.destroyAllWindows() # removes video from memory
cv.waitKey(1)
print(counter)
```

```python
######## Translation
def translate(img, x, y):
    transMat = np.float32([[1,0,x],[0,1,y]])
    dimensions = (img.shape[1], img.shape[0])
    return cv.warpAffine(img, transMat, dimensions)

translated = translate(img, -100, 100)
cv.imshow('Translated', translated)
cv.waitKey(0)
cv.destroyAllWindows() 
cv.waitKey(1)
```

```python
# Rotation
def rotate(img, angle, rotPoint=None):
    (height,width) = img.shape[:2]

    if rotPoint is None:
        rotPoint = (width//2,height//2)
    
    rotMat = cv.getRotationMatrix2D(rotPoint, angle, 1)
    dimensions = (width,height)

    return cv.warpAffine(img, rotMat, dimensions)

rotated = rotate(img, 0)
cv.imshow('Rotated', rotated)
cv.waitKey(0)
cv.destroyAllWindows() 
cv.waitKey(1)
```

```python
# Flipping
flip = cv.flip(img, 1 )
cv.imshow('Flip', flip)
cv.waitKey(0)
cv.destroyAllWindows() 
cv.waitKey(1)
```
